```
import json
import requests
from bs4 import BeautifulSoup

import pandas as pd
import matplotlib.pyplot as plt

urlArray = ["https://www.mohfw.gov.in/","https://www.worldometers.info/coronavirus/"]
def parse_service_india(*args):
    res = requests.get(args[0])
    soup = BeautifulSoup(res.content,'lxml')
    table_div = soup.find('div' , {'id': 'cases'})
    table = table_div.find('table')
    df = pd.read_html(str(table))

    new_df = df[0]
    return new_df

def parse_service_world(*args):
    res = requests.get(args[0])
    soup = BeautifulSoup(res.content,'lxml')
    table = soup.find_all('table')[0] 
    df = pd.read_html(str(table))

    new_df = df[0]
    return new_df


def in_live(url):
    df = parse_service_india(url)
    # print(df)
    new_df = df[:-2]
    # print(new_df)
    labels = new_df["Name of State / UT"].to_list()
    confirmed_indian_case = new_df["Total Confirmed cases (Indian National)"].to_list()
    confirmed_foreig_case = new_df["Total Confirmed cases ( Foreign National )"].to_list()
    discharged = new_df["Cured/Discharged/Migrated"].to_list()
    death = new_df["Death"].to_list()
    return labels, confirmed_indian_case, confirmed_foreig_case, discharged, death

def world_data(url):
    df = parse_service_world(url)
    new_df = df[:-1].fillna(0)

    country_labels = new_df["Country,Other"].to_list()
    total_cases = new_df["TotalCases"].to_list()
    new_cases = new_df["NewCases"].to_list()
    total_deaths = new_df["TotalDeaths"].to_list()
    new_deaths = new_df["NewDeaths"].to_list()
    total_recovered = new_df["TotalRecovered"].to_list()
    active_cases = new_df["ActiveCases"].to_list()
    critical = new_df["Serious,Critical"].to_list()
    
    return country_labels,total_cases,new_cases,total_deaths,new_deaths,total_recovered,active_cases,critical 

c_labels,total_cases,new_cases,total_deaths,new_deaths,total_recovered,active_cases,critical = world_data(urlArray[1])
#print(c_labels,total_cases,new_cases,total_deaths,new_deaths,total_recovered,active_cases,critical)

labels, confirmed_indian_case, confirmed_foreig_case, discharged, death  = in_live(urlArray[0])
#print(labels, confirmed_indian_case, confirmed_foreig_case, discharged, death)

plt.barh(labels, confirmed_indian_case)
plt.xlabel("Confirmed Cases") 
plt.ylabel("State") 
plt.show() 
```

```
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(
    x=labels,
    y=confirmed_indian_case,
    name='+ve cases (Indian National)',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=labels,
    y=confirmed_foreig_case,
    name='+ve cases ( Foreign National )',
    marker_color='DarkBlue'
))
fig.add_trace(go.Bar(
    x=labels,
    y=death,
    name='Death',
    marker_color='Red'
))
fig.add_trace(go.Bar(
    x=labels,
    y=discharged,
    name='cured',
    marker_color='Green'
))

fig.update_layout(barmode='group', xaxis_tickangle=-60)
fig.show()
```

```
fig = go.Figure(go.Bar(x=c_labels, y=total_cases, name='Total Cases', marker_color='Black'))
fig.add_trace(go.Bar(x=c_labels, y=new_cases, name='New Cases', marker_color='indianred'))
fig.add_trace(go.Bar(x=c_labels, y=total_deaths, name='Death', marker_color='Red'))
fig.add_trace(go.Bar(x=c_labels, y=total_recovered, name='cured', marker_color='Green'))
fig.add_trace(go.Bar(x=c_labels, y=new_deaths, name='New Deaths', marker_color='rgb(204, 0, 204)'))
fig.add_trace(go.Bar(x=c_labels, y=active_cases, name='Active Cases', marker_color='rgb(255, 153, 0)'))
fig.add_trace(go.Bar(x=c_labels, y=critical, name='Critical', marker_color='rgb(204, 153, 0)'))

fig.update_layout(barmode='stack', xaxis_tickangle=-60, xaxis={'categoryorder':'array', 'categoryarray':['d','a','c','b']})

fig.show()
```
